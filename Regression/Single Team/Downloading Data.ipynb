```python
import pandas as pd
import os
import unicodedata
from io import StringIO
import requests
```

```python
# Downloading the yahoo (Moneyball Stats)
tableY = pd.read_html('https://sports.yahoo.com/mlb/teams/atlanta/stats/')
dataY = tableY[0]
```

```python
# Fixes player names (accents)
dataY['Player'] = (
    dataY['Player']
    .apply(lambda x: unicodedata.normalize('NFKD', x)
    .encode('ascii', 'ignore')
    .decode('utf-8'))
)
```

```python
dataY['Score'] = round(((dataY['H'] + dataY['BB'] - dataY['CS']) + ((dataY['2B'] * 2 + dataY['3B'] * 3 + dataY['HR'] * 4 + (dataY['H'] - dataY['2B'] - dataY['3B'] - dataY['HR'])) + .7 * dataY['SB'])) / (dataY['AB'] + dataY['BB'] + dataY['CS']),3)
```

```python
dataY.to_csv('Braves_Yahoo.csv', index=False)
os.startfile('Braves_Yahoo.csv')
```
