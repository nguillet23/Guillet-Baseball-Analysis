```python
import pandas as pd
```

```python
# Import MLB Data
tables = pd.read_html("https://sports.yahoo.com/mlb/teams/atlanta/stats/?guccounter=1&guce_referrer=aHR0cHM6Ly93d3cuZ29vZ2xlLmNvbS8&guce_referrer_sig=AQAAAKUaBNRUkewwkwceQASnKiUAqrzRcvlr19rcsdZK8bMOXm0jO4LAqGY8Ywq8ADRgvSf_WHEcIrKIMZFDSljt9uy-Gmf5EOkZ7gQzfkMN9BHwU3dQVCMoUHlCn2Z4bdMj_mEBQQebWUJdXoZXFbSSJjFRA_Dm2bfhkIy9iaILG4QA")
data = tables[0]
```

```python
# Cleaning the Data
data = data.drop(['G','R','RBI','K','AVG','OBP','SLG','OPS'], axis=1)
```

```python
# Calculate total bases
data['Total Bases'] = (
    (data['H'] - data['2B'] - data['3B'] - data['HR']) +
    2 * data['2B'] +
    3 * data['3B'] +
    4 * data['HR']
)
```

```python
# Calculating of "Projected Runs"
numerator = (data['H'] + data['BB'] - data['CS']) * (data['Total Bases'] + 0.7 * data['SB'])
denominator = data['AB'] + data['BB'] + data['CS']

data['Projected Runs'] = numerator / denominator.replace(0, pd.NA)
data['Projected Runs'] = data['Projected Runs'].fillna(0)  
data['Projected Runs'] = data['Projected Runs'].round().astype(int)
```

```python
data.to_csv('Data.csv', index=False)
```
