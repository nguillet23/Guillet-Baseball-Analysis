```python
import requests
import pandas as pd
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
url = "https://baseballsavant.mlb.com/team/144?view=statcast&nav=hitting&season=2025"
response = requests.get(url, headers=headers)
tables = pd.read_html(response.text)
sprinting = tables[6]
sprinting = sprinting.drop(columns=[('Age'),('Lg Rank'),('Pos Rank'),('Age Rank'),('% Rank')])
```

```python
sprinting.to_csv('Sprinting.csv', index=False)
```
