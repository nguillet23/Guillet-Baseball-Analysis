```python
# Importing Libraries
import pandas as pd
```

```python
# Importing Data
url = "https://sports.yahoo.com/mlb/teams/atlanta/stats/"
tables = pd.read_html(url)
data = tables[0]
```

```python
# Cleaning/dropping unwanted columns
cols_to_drop = ['G', 'R', 'RBI', 'K', 'AVG', 'OBP', 'SLG', 'OPS','H', '2B', '3B', 'HR', 'BB', 'AB','Final','E']
data = data.drop([col for col in cols_to_drop if col in data.columns], axis=1)
```

```python
# Saving Data to CSV File
data.to_csv("Stolen Bases.csv", index=False)
```
