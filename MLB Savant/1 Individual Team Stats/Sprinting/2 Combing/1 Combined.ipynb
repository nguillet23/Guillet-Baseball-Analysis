```python
# Importing Libraries
import requests
import pandas as pd
```

```python
# Importing Data
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
url_Sav = "https://baseballsavant.mlb.com/team/137?view=statcast&nav=hitting&season=2025"
response = requests.get(url_Sav, headers=headers)
tables = pd.read_html(response.text)
sprinting = tables[6]

# Importing Data
url_Yah = "https://sports.yahoo.com/mlb/teams/san-francisco/stats/"
tables = pd.read_html(url_Yah)
SB = tables[0]
```

```python
# For loop to collect matched players' stats
merged_rows = []

for i in range(len(sprinting)):
    sprint_row = sprinting.iloc[i]
    sprint_name = sprint_row['Player']
    
    # Look for exact match in SB
    matching_rows = SB[SB['Player'] == sprint_name]
    if not matching_rows.empty:
        sb_row = matching_rows.iloc[0]
        
        # Combine sprinting and SB row data
        combined = pd.concat([sprint_row, sb_row.drop(labels='Player')])
        merged_rows.append(combined)

# Create DataFrame from list of merged rows
merged_df = pd.DataFrame(merged_rows)
```

```python
merged_df = merged_df.drop(columns=[('Age'),('Pos Rank'),('Age Rank'),('G'),('G'),('AB'),('R'),('H'),('2B'),('3B'),('HR'),('RBI'),('BB'),('K'),('AVG'),('OBP'),('SLG'),('OPS'),])
```

```python
# Save to CSV
merged_df.to_csv("Combined.csv", index=False)
```
