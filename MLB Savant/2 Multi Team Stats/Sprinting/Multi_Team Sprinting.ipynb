```python
import requests
import pandas as pd
from time import sleep
```

```python
# Yahoo and Savant team mappings
team_info = {
    108: "la-angels", 109: "arizona", 144: "atlanta", 110: "baltimore", 111: "boston",
    112: "chi-cubs", 113: "cincinnati", 114: "cleveland", 115: "colorado", 116: "detroit",
    117: "houston", 118: "kansas-city", 119: "la-dodgers", 121: "miami", 133: "milwaukee",
    134: "minnesota", 121: "ny-mets", 147: "ny-yankees", 133: "oakland", 143: "philadelphia",
    134: "pittsburgh", 135: "san-diego", 136: "seattle", 137: "san-francisco", 138: "st-louis", 139: "tampa-bay", 140: "texas",
    141: "toronto", 120: "washington"
}

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Initialize final combined DataFrame
final_merged_df = pd.DataFrame()
```

```python
for sav_id, yah_abbr in team_info.items():
    try:
        print(f"Processing: {yah_abbr}")

        # Baseball Savant URL
        url_Sav = f"https://baseballsavant.mlb.com/team/{sav_id}?view=statcast&nav=hitting&season=2025"
        response_sav = requests.get(url_Sav, headers=headers)
        tables_sav = pd.read_html(response_sav.text)

        if len(tables_sav) < 7:
            print(f"Skipping {yah_abbr} due to missing Statcast table.")
            continue

        sprinting = tables_sav[6]

        # Yahoo Sports URL
        url_Yah = f"https://sports.yahoo.com/mlb/teams/{yah_abbr}/stats/"
        tables_yah = pd.read_html(url_Yah)

        if len(tables_yah) == 0:
            print(f"Skipping {yah_abbr} due to missing Yahoo table.")
            continue

        SB = tables_yah[0]

        # Match rows by exact player name
        merged_rows = []
        for i in range(len(sprinting)):
            sprint_row = sprinting.iloc[i]
            sprint_name = sprint_row['Player']

            matching_rows = SB[SB['Player'] == sprint_name]
            if not matching_rows.empty:
                sb_row = matching_rows.iloc[0]
                combined = pd.concat([sprint_row, sb_row.drop(labels='Player')])
                merged_rows.append(combined)

        if merged_rows:
            merged_df = pd.DataFrame(merged_rows)
            merged_df["Team"] = yah_abbr.upper()

            # Drop unnecessary columns
            to_drop = ['Age', 'Pos Rank', 'Age Rank', 'G', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI', 'BB', 'K', 'AVG', 'OBP', 'SLG', 'OPS']
            merged_df = merged_df.drop(columns=[col for col in to_drop if col in merged_df.columns], errors='ignore')
            final_merged_df = pd.concat([final_merged_df, merged_df], ignore_index=True)

        sleep(1)

    except Exception as e:
        print(f"Error processing {yah_abbr}: {e}")
```

```python
# Save to CSV
final_merged_df.to_csv("Multi_Team Sprinting.csv", index=False)
```
